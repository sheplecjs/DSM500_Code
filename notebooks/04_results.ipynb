```python
import re
import pandas as pd
```

```python
def get_results(file: str) -> pd.DataFrame:
    """Grabs results from text file and organizes into dataframe.

    Args:
        file (str): path to the results file.

    Returns:
        pd.DataFrame
    """

    p = []

    reg = re.compile("[0-9.]+")

    with open(file) as f:
        lines = f.readlines()

    for l, s in zip(lines[::3], lines[1::3]):
        parsed = l.split("_")
        dataset = parsed[0]
        try:
            t = int(parsed[1])
            trunc = t
            model = parsed[2] + parsed[3]
        except ValueError:
            trunc = 1
            model = parsed[1] + parsed[2]

        scores = reg.findall(s)
        mse = scores[0][:5]
        mae = scores[1][:5]

        p.append([dataset, trunc, model, mse, mae])

    return pd.DataFrame(p, columns=["data", "truncated", "model", "mse", "mae"])
```

```python
results = get_results("result_old.txt")
```

```python
r = results[results["truncated"].isin([1, 4])]

r
```
