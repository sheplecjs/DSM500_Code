```python
# for sdm model training
# iterate over models and datasets (100%, 80%, 60%, 40% truncated)
for data, name in zip(datasets, data_names):
    cmd = f"python -u ../run.py\
    --is_training 1 \
    --root_path datasets/ \
    --data_path {data} \
    --model_id {name} \
    --model Transformer_sdm \
    --data custom \
    --features S \
    --seq_len 48 \
    --label_len 24 \
    --pred_len 24 \
    --e_layers 3 \
    --d_layers 3 \
    --factor 1 \
    --enc_in 1 \
    --dec_in 1 \
    --c_out 1 \
    --des 'Exp' \
    --itr 1 \
    --n_heads 1"

    subprocess.run(cmd, shell=True)
```
